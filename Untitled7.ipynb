Load Dataset

```python
df = spark.read.csv("logistics_data.csv", header=True, inferSchema=True)
df.show(5)
```

Data Aggregation (Group By)

```python
from pyspark.sql.functions import avg

df_group = df.groupBy("origin").agg(avg("cost").alias("avg_cost"))

df_group.show()
```

Crate Date Column

```python
from pyspark.sql.functions import current_date

df = df.withColumn("date", current_date())

df.show(5)
```

Window Functions By Date Wise
1.Running Total

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum

window_spec = Window.partitionBy("origin").orderBy("date")

df = df.withColumn("running_total", sum("cost").over(window_spec))

df.show()
```

Window Functions By shipment_id Wise
1.Running Total

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum

window_spec = Window.partitionBy("origin").orderBy("shipment_id")

df = df.withColumn("running_total", sum("cost").over(window_spec))

df.show()
```

2.Rolling Average By Date Wise

```python
from pyspark.sql.functions import avg

window_spec2 = Window.partitionBy("origin").orderBy("date").rowsBetween(-2, 0)

df = df.withColumn("rolling_avg", avg("cost").over(window_spec2))

df.show()
```

Rolling Average By shipment_id Wise

```python
from pyspark.sql.functions import avg

window_spec2 = Window.partitionBy("origin").orderBy("shipment_id").rowsBetween(-2, 0)

df = df.withColumn("rolling_avg", avg("cost").over(window_spec2))

df.show()
```

Dag & Execution Plan (Directed Acyclic Graph)

```python
df.explain(True)
```

Optimization
1.Reduce Shuffles

```python
df = df.repartition(2)
```

2.Cache Data

```python
df.cache()
```

Selected Required Columns Only

```python
df = df.select("origin", "cost", "date","shipment_id")
```
