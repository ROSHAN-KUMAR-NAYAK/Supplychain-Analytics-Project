Start Spark Session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("LogiScale Project Week 1") \
    .getOrCreate()

print("Spark Session Created")
```

Create Sample Logistics Dataset

```python
import pandas as pd
import numpy as np

# Create large dataset (1 Thousand  rows)
data = {
    "shipment_id": np.arange(100000),
    "origin": np.random.choice(["India", "USA", "China", "Germany"], 100000),
    "destination": np.random.choice(["India", "USA", "China", "Germany"], 100000),
    "distance_km": np.random.randint(50, 10000, 100000),
    "cost": np.random.randint(100, 5000, 100000)
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("logistics_data.csv", index=False)
```

Download The Logistics Data Set

```python
from google.colab import files

files.download("logistics_data.csv")
```

Load Data Using Pandas(Benchmark)

```python
import pandas as pd
import time

start = time.time()

pandas_df = pd.read_csv("logistics_data.csv")

end = time.time()

print("Pandas Load Time:", end - start, "seconds")
```

Load Data Using Spark

```python
start = time.time()

spark_df = spark.read.csv("logistics_data.csv", header=True, inferSchema=True)

end = time.time()

print("Spark Load Time:", end - start, "seconds")
```

Show Data

```python
spark_df.show(5)
```

Basic Analysis (Spark)

```python
spark_df.printSchema()

spark_df.groupBy("origin").count().show()

spark_df.select("origin", "cost").show(5)
```

Convert To Perquet (Important For Big Data)

```python
spark_df.write.parquet("logistics_parquet")
```

Read Again

```python
parquet_df = spark.read.parquet("logistics_parquet")
parquet_df.show(5)
```
